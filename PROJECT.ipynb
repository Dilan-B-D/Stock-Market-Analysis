**Introduction to Stock Market Analysis**


Stock market analysis is the process of evaluating financial markets, individual stocks, and economic trends to make informed investment decisions. It helps traders, investors, and analysts understand market behavior, predict price movements, and optimize portfolios for profitability.
There are two primary methods of stock market analysis:
- Fundamental Analysis – Examines financial statements, earnings reports, industry trends, and macroeconomic factors to determine a stock's intrinsic value.
- Technical Analysis – Uses historical price data, charts, and indicators like moving averages, Bollinger Bands, and RSI to predict future price movements.

Additionally, quantitative analysis and machine learning have emerged as modern techniques to enhance stock predictions using big data. Understanding these strategies allows investors to mitigate risks and capitalize on opportunities in a dynamic financial landscape.

**Stock Market Analysis Using Line, Bar, and Scatter Graphs
Stock market analysis uses different types of visualizations to understand trends, compare performance, and identify patterns. Here’s how line graphs, bar charts, and scatter plots help in analyzing financial data**


```python
import matplotlib.pyplot as plt

# Stock portfolio data (example percentage holdings)
stocks = ["Apple (AAPL)", "Tesla (TSLA)", "Microsoft (MSFT)", "Amazon (AMZN)", "Google (GOOGL)"]
values = [30, 20, 25, 15, 10]  # Portfolio percentage allocations

# Create pie chart
plt.figure(figsize=(8, 8))
plt.pie(values, labels=stocks, autopct="%1.1f%%", colors=["blue", "red", "green", "purple", "orange"], startangle=140)
plt.title("Stock Portfolio Allocation")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("stocks.csv")


df["Date"] = pd.to_datetime(df["Date"])


tickers = df["Ticker"].unique()


for ticker in tickers:
    stock_data = df[df["Ticker"] == ticker]

# plotting line graph
    plt.figure(figsize=(12, 6))
    plt.plot(stock_data["Date"], stock_data["Close"], label="Closing Price", color="red")
    plt.title(f"{ticker} Closing Prices Over Time")
    plt.xlabel("Date")
    plt.ylabel("Closing Price")
    plt.legend()
    plt.grid()
    plt.show()

# plotting bar graph
    plt.figure(figsize=(12, 6))
    plt.bar(stock_data["Date"], stock_data["Volume"], label="Volume", color="blue")
    plt.title(f"{ticker} Trading Volume Over Time")
    plt.xlabel("Date")
    plt.ylabel("Volume")
    plt.legend()
    plt.grid()
    plt.show()
# plotting line graph
    plt.figure(figsize=(12, 6))
    plt.plot(stock_data["Date"], stock_data["Open"], label="Closing Price", color="green")
    plt.title(f"{ticker} Opening Price Over Time")
    plt.xlabel("Date")
    plt.ylabel("Opening Price")
    plt.legend()
    plt.grid()
    plt.show()
```

**Heat Map in Stock Market Analysis**


**A heat map is a visual representation of data where values are represented by varying colors, helping analysts quickly identify patterns, correlations, and outliers in large datasets. In stock market analysis, heat maps are used for tracking market performance, sector movements, and price correlations.**


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# uploading the csv file
df = pd.read_csv("stocks.csv")


pivot_data = df.pivot(index="Date", columns="Ticker", values="Close")
pivot_data.index = pd.to_datetime(pivot_data.index)

# creating correlation matrix
correlation_matrix = pivot_data.corr()

plt.figure(figsize=(10, 8))
# colormapping
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Stock Closing Prices")
plt.show()
```

**Preprocessing in Stock Market Analysis**


Preprocessing is a crucial step in stock market analysis where raw data is cleaned, transformed, and prepared for analysis. It ensures accuracy and efficiency in predicting trends using technical indicators or machine learning models.
Key Preprocessing Steps in Stock Market Data
- Data Cleaning – Removing missing or duplicate values, handling inconsistent data entries.
- Feature Engineering – Creating new meaningful features (e.g., moving averages, volatility indicators).
- Normalization & Scaling – Standardizing stock prices and volumes for better model performance.
- Time-Series Formatting – Converting data into structured formats suitable for time-series analysis.
- Handling Missing Values – Using interpolation or forward-fill methods for missing data points.




```python
# preprocessing the csv file
import pandas as pd
from sklearn.preprocessing import MinMaxScaler


file_path = "stocks.csv"
df = pd.read_csv(file_path)

# Identifying the null values/ Handling the missing values
print("Missing values before cleaning:")
print(df.isnull().sum())

#columns of csv file
numerical_columns = ["Open", "High", "Low", "Close", "Adj Close", "Volume"]
df[numerical_columns] = df[numerical_columns].fillna(df[numerical_columns].mean())

# Scaling the data
scaler = MinMaxScaler()
df[numerical_columns] = scaler.fit_transform(df[numerical_columns])

# Replace the null values
df.columns = [col.lower().replace(" ", "_") for col in df.columns]


print("\nData after preprocessing:")
print(df.head())

# importing the processed data
df.to_csv("stocks_preprocessed.csv", index=False)
```

**Principal Component Analysis (PCA) in Stock Market Analysis**


Principal Component Analysis (PCA) is a dimensionality reduction technique used in data analysis to transform complex datasets into a smaller number of significant variables while retaining essential patterns. In stock market analysis, PCA is useful for reducing the number of features while preserving the most impactful information.

Why Use PCA in Stock Market Analysis?
✔ Feature Reduction: Stock market datasets contain multiple indicators; PCA helps condense them.
✔ Pattern Identification: Extracts key trends from correlated variables like volume, price, volatility.
✔ Noise Reduction: Removes redundant data to improve prediction models.
✔ Visualization: Helps in representing multi-variable financial data in fewer dimensions.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler

# Load dataset
df = pd.read_csv("stocks.csv")

# Selecting numerical features
features = ['Open', 'High', 'Low', 'Close', 'Volume']
X = df[features].dropna()

# Standardizing features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Feature selection using PCA
pca = PCA(n_components=2)  # Reduce to 2 principal components
X_pca = pca.fit_transform(X_scaled)

# Visualizing PCA components
plt.figure(figsize=(6,4))
plt.scatter(X_pca[:, 0], X_pca[:, 1], alpha=0.5)
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA Feature Visualization")
plt.show()

# Feature importance using Random Forest
y = df['Close'].dropna()  # Assuming 'Close' price is the target variable
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_scaled, y)

# Plotting feature importance
importance = rf.feature_importances_
plt.figure(figsize=(6,4))
plt.barh(features, importance, color='skyblue')
plt.xlabel("Feature Importance Score")
plt.title("Feature Importance using Random Forest")
plt.show()
```

```python
# Install necessary libraries
!pip install pandas numpy matplotlib yfinance seaborn --quiet

#Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf

# Download stock data
ticker = "GOOG"  # Example: Apple Inc.
df = yf.download(ticker, start="2020-01-01", end="2025-01-01")

# Visualize stock price trends
plt.figure(figsize=(12, 6))
plt.plot(df['Close'], label=f'{ticker} Closing Price', color='blue')
plt.title(f'{ticker} Stock Price Trend')
plt.xlabel('Date')
plt.ylabel('Stock Price (USD)')
plt.legend()
plt.grid(True)
plt.show()

# Calculate Moving Averages
df['SMA_50'] = df['Close'].rolling(window=50).mean()
df['SMA_200'] = df['Close'].rolling(window=200).mean()

#  Plot Moving Averages
plt.figure(figsize=(12, 6))
plt.plot(df['Close'], label='Close Price', color='blue', alpha=0.6)
plt.plot(df['SMA_50'], label='50-Day SMA', color='green')
plt.plot(df['SMA_200'], label='200-Day SMA', color='red')
plt.title(f'{ticker} Stock Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()

# Calculate Relative Strength Index (RSI)
def calculate_rsi(data, window=14):
    delta = data['Close'].diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()
    rs = gain / loss
    data['RSI'] = 100 - (100 / (1 + rs))

calculate_rsi(df)

# Plot RSI
plt.figure(figsize=(12, 6))
plt.plot(df['RSI'], label='RSI', color='purple')
plt.axhline(30, linestyle='--', color='red')  # Oversold line
plt.axhline(70, linestyle='--', color='green')  # Overbought line
plt.title(f'{ticker} Relative Strength Index (RSI)')
plt.xlabel('Date')
plt.ylabel('RSI Value')
plt.legend()
plt.grid(True)
plt.show()

# Display last few rows of dataset
print(df.tail())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load stock data
df = pd.read_csv("stocks.csv")

# Define up/down days based on closing prices
df['trend'] = df['Close'].diff().apply(lambda x: 'Up' if x > 0 else 'Down')

# Count occurrences
color_counts = df['trend'].value_counts()

# Plotting the results
plt.figure(figsize=(6, 4))
color_counts.plot(kind='bar', color=['red', 'green'])
plt.xlabel("Market Trend")
plt.ylabel("Count of Days")
plt.title("Stock Market Up and Down Days Count")
plt.xticks(rotation=0)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Sample dataset with missing values
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [np.nan, 2, 3, np.nan, 5],
    'C': ['X', 'Y', 'Z', np.nan, 'X']
}

df = pd.DataFrame(data)
print("Original Dataset:")
print(df)

# Handling missing values
df['A'].fillna(df['A'].mean() , inplace=True)
df.fillna(df['B'].median() , inplace=True)
df['C'].fillna(df['C'].mode() , inplace=True)
print("\nDataset after filling missing values:")
print(df)

# Data transformation
df['A'] = df['A'] * 10  # Scaling column A
df['B'] = np.log1p(df['B'])  # Applying log transformation
df['C'] = df['C'].str.lower()
plt.savefig('x.png')  # Transforming text column

print("\nTransformed Dataset:")
print(df)
```
